## Easemytrip - Web Scrapping 
#### date - 21/june/2022

```python
# let's import all required libraries
import selenium
from selenium import webdriver
import pandas as pd
import time
import warnings
warnings.filterwarnings('ignore')
from bs4 import BeautifulSoup
import requests

from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException,ElementNotInteractableException
```

```python
driver=webdriver.Chrome(r"/Users/vikram/Downloads/chromedriver")
time.sleep(2)
url="https://www.easemytrip.com/"
driver.get(url)#enter to olx.in through chrome browser
```

```python
driver.execute_script("window.scrollBy(0,800)")
```

```python
urls=[]
for i in driver.find_elements_by_xpath("//a[@class='srch_btn']"):
    urls.append(i.get_attribute("href"))
```

```python
len(urls)
```

```python
airline = [] 
dep_time = []
arrival_time = []
source = []
destination = []
stops= [] 
duration= [] 
price= []
```

```python
# date = 21 june 2022
for i in urls:
    driver.get(i)#loading the url on web driver
    time.sleep(4) # giving the webpage 4 seconds to load
    
    name=driver.find_elements_by_xpath("//span[@class='txt-r4 ng-binding']")
    for i in name:
        airline.append(i.text)
        
    dep=driver.find_elements_by_xpath("//div[@class='col-md-2 col-sm-2 col-xs-4 top5 ']//span[1]")
    for i in dep:
        dep_time.append(i.text) 
        
    arv = driver.find_elements_by_xpath("//div[@class='col-md-2 col-sm-2 col-xs-3 top5']//span[1]")
    for i in arv:
        arrival_time.append(i.text)
    
    sou=driver.find_elements_by_xpath("//div[@class='col-md-2 col-sm-2 col-xs-4 top5 ']//span[2]")
    for i in sou:
        source.append(i.text)
        
    des = driver.find_elements_by_xpath("//div[@class='col-md-2 col-sm-2 col-xs-3 top5']//span[2]")
    for i in des:
        destination.append(i.text)
     
    tm = driver.find_elements_by_xpath("//div[@class='col-md-2 col-sm-2 col-xs-5 non-st']//span[1]")
    for i in tm:
        duration.append(i.text)
        
    sts = driver.find_elements_by_xpath("//div[@class='col-md-2 col-sm-2 col-xs-5 non-st']//span[2]")
    for i in sts:
        stops.append(i.text)

    rate = driver.find_elements_by_xpath("//div[@class='col-md-8 col-sm-8 col-xs-9 txt-r6-n exPrc ng-binding']")
    for i in rate:
        price.append(i.text)
    
```

```python
#DATA FRAMEING
Flights=pd.DataFrame({})
Flights['Airline'] = airline
Flights['Source'] = source
Flights['Destination'] = destination
Flights['Depature Time'] = dep_time
Flights['Arrival Time'] = arrival_time
Flights['Stops'] = stops
Flights['Duration'] = duration[:1352]
Flights['Price'] = price

Flights # for 21 June
```

```python
date = [] # adding date to dataframe

for i in range(1352):
    date.append("21/06/2022")
date    
```

```python
Flights['Date'] = date
Flights
```

```python
driver.close() #closing driver to open new for another date
```
